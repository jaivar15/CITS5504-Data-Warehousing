### Data Cleaning

```python
#drop the first column - id 
mobile_data = mobile_data.drop('id', axis = 'columns')
```

```python
mobile_data.isnull().sum().sum()
```

```python
def rename_binary_attributes(data, attribute):
    for index in attribute:  
        data[attribute] = data[attribute].replace(['Yes', 'YES', 'has','Has','yes'],1)
        data[attribute] = data[attribute].replace(['NO', 'not', 'Not','No','no'],0)
    return data
mobile_data = rename_binary_attributes(mobile_data, ['blue','dual_sim','three_g','wifi'])
```

```python
def drop_column(dataset, column_name):
    return dataset.drop(column_name, axis = 'columns')
```
